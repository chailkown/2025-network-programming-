# 컴퓨터 네트워크 개요

컴퓨터 네트워크는 정보와 자원 공유를 위해 링크로 연결된 노드(컴퓨터)들의 집합이다.

## 네트워크 구성 요소

1. **메시지(Message)**: 전송할 데이터 (패킷으로 구성)
2. **송신자(Sender)**: 메시지를 보내는 장치
3. **수신자(Receiver)**: 메시지를 받는 장치
4. **전송매체(Medium)**: 물리적인 전송 경로 (케이블, 무선 등)
5. **프로토콜(Protocol)**: 데이터 통신을 위한 규칙들의 집합


## 프로토콜의 구조

프로토콜은 계층 구조로 구성되어 있다.

### 프로토콜의 3가지 요소
- **구문(Syntax)**: 전송되는 데이터의 구조 및 형식
- **의미(Semantics)**: 전송되는 데이터의 의미
- **순서(Timing)**: 데이터의 전송 순서

### 계층 구조의 장점
- 복잡한 시스템을 계층별로 나누어 모듈화
- 모듈 사이의 인터페이스만 유지하면 유지관리, 수정 용이
- (N) 계층은 (N-1) 계층의 서비스를 받아 (N+1) 계층에게 서비스 제공


## 각 계층의 상세 기능

### 물리계층 (1계층)
- 실제 장치를 연결하여 데이터를 전송하기 위한 기능과 처리절차 규정
- 비트의 표현, 데이터 속도, 비트의 동기화

### 데이터링크계층 (2계층)
- 직접 연결된 노드 사이 신뢰성 있는 데이터 전송 책임
- 프레임 구성, 물리주소 지정, 흐름제어, 오류제어, 접근제어

### 네트워크계층 (3계층)
- 발신지에서 목적지까지 패킷을 전달하는 책임
- 논리주소 지정, 라우팅

### 전송계층 (4계층)
- 전체 메시지의 프로세스 대 프로세스 전달에 대한 책임
- 포트주소 지정, 분할과 재조립, 연결제어, 흐름제어, 오류제어


### 세션계층 (5계층)
- 응용 프로세스가 대화 단위로 통신을 관리하기 위한 방법 제공
- 논리적 연결의 설정, 유지, 종료, 복구 등

### 표현계층 (6계층)
- 시스템 간에 교환되는 데이터의 형식화와 표현을 담당
- 데이터 암호화/복호화, 데이터 압축/확장

### 응용계층 (7계층)
- 최종 사용자(응용프로그램)에게 서비스를 제공
- 원격로그인(Telnet), 파일전송(FTP), 우편서비스(E-mail), WWW 서비스


# TCP/IP 프로토콜

인터넷에서 사용하는 4계층 프로토콜이다.

## TCP/IP 등장 배경
- 1968년 미국 국방부 고등연구계획국(DARPA)에서 알파넷(ARPAnet) 구축
- 1973년 전송제어프로토콜(TCP) 개발
- 1982년 TCP/IP 프로토콜 탄생

## TCP/IP 특징
- 인터네트워킹을 위한 프로토콜
- 여러 프로토콜의 그룹
- TCP와 IP가 가장 중요한 계층 프로토콜


## TCP/IP 4계층 상세

### 1. 네트워크 인터페이스 계층
- 네트워크 어댑터(NIC)를 위한 장치 드라이버 및 소프트웨어
- IP 데이터그램을 서브네트워크가 이해할 수 있는 패킷으로 변환

### 2. 네트워크 계층
- TCP/UDP 레벨의 전송 데이터를 데이터그램으로 구성하여 전송
- 경로 배정 기능
- 대표 프로토콜: IP

### 3. 전송 계층
- 오류제어와 흐름제어를 이용하여 프로세스 사이의 전송을 책임
- 프로토콜: TCP, UDP

### 4. 응용 계층
- 최종 사용자에게 서비스를 제공
- 프로토콜: WWW, e-mail, FTP, Telnet 등


## TCP와 UDP 헤더 구조

### TCP 헤더 (20바이트)
- 양방향 통신
- 발신/수신 포트번호
- 순서번호: 임의의 수~4G까지 0-1씩 증가
- 확인번호: 기대하는 다음 수신번호
- 윈도우크기: 수신버퍼 여유용량 크기
- 각종 플래그들: 흐름제어에 활용

### UDP 헤더 (8바이트)
- 발신포트/수신포트번호: 각 16비트 (2B+2B)
- UDP 크기: 헤더와 데이터 포함한 바이트수 (2B)
- 체크섬: 오류 검출코드 (2B)


# IP 주소

인터넷에서 호스트를 구별하기 위해 사용하는 글로벌 주소다.

## IP 주소가 필요한 이유
- 서브네트워크(LAN) 내에서 호스트 식별은 물리주소(MAC) 사용
- 인터넷은 서브네트워크를 연결하여 구성된 네트워크
- 물리주소 체계가 다른 경우, 상호 식별을 위해 공통으로 사용하는 글로벌 주소 필요

## IPv4와 IPv6
- **IPv4**: 32비트 (예: 192.168.0.10)
- **IPv6**: 128비트 (예: 3FFE:F200:0234:AB00:0123:4567:8901:ABCD)


## IP 주소 구성

### 네트워크 주소 + 호스트 주소
- **네트워크 주소 부분**: 서브네트워크를 식별하기 위해 사용
- **호스트 주소 부분**: 네트워크 내에서 호스트 식별을 위해 사용

예시: `192.168.123.132`
- 네트워크 주소: `192.168.123`
- 호스트 주소: `132`

### 서브넷 마스크
- 네트워크 주소와 호스트 주소를 구분하기 위해 사용
- 네트워크 주소로 사용되는 비트는 1, 호스트 주소로 사용되는 부분은 0으로 표시
- 예: `255.255.255.0`


## IP 주소 클래스

IP 주소의 처음 5비트를 사용하여 클래스를 구분한다.

| 클래스 | 범위 | 예시 |
|--------|------|------|
| A | 1.0.0.0 ~ 126.255.255.255 | 10.52.36.11 |
| B | 128.0.0.0 ~ 191.255.255.255 | 172.1.52.63 |
| C | 192.0.0.0 ~ 223.255.255.255 | 192.168.123.132 |
| D | 224.0.0.0 ~ 239.255.255.255 | 멀티캐스트 |
| E | 240.0.0.0 ~ 255.255.255.255 | 유보 |


# CIDR (Classless Inter-Domain Routing)

클래스 기반 IP 주소 체계의 비효율성을 개선하기 위해 1993년에 도입된 IP 주소 할당 및 라우팅 방법이다.

## CIDR 표기법
- 형태: `IP 주소/프리픽스`
- 예시: `192.168.1.0/24`
  - 처음 24비트가 네트워크 부분
  - 나머지 8비트가 호스트 부분
  - 호스트는 2^8 = 256개까지 사용 가능
  - 단, 0은 네트워크 주소, 255는 브로드캐스트 주소로 사용

## CIDR 장점
- IP 주소 효율화
- 라우팅 효율성 증가
- 가변길이 서브넷마스크 사용 가능


# MAC 주소

미디어 접근 제어(Media Access Control) 주소로, LAN에서 네트워크 인터페이스 카드를 구분하는 유일한 식별자다.

## MAC 주소 용도
- **네트워크 제어**: 라우터나 스위치 같은 장비가 MAC 주소를 이용해 패킷을 목적지로 전송
- **보안 필터링**: 무선 공유기 등에서 특정 MAC 주소를 허용하거나 차단
- **장치 식별**: CCTV와 같은 장비에서 특정 장치를 고유하게 식별

## MAC 주소 확인 방법
Windows 명령 프롬프트에서:
```bash
ipconfig /all
```


# 네트워크 장비

## 유선 LAN
- **HUB**: 서로 다른 MAC주소를 가진 컴퓨터끼리 통신 (1:1)
  - 저가, 느림
- **Switching HUB**: 스위치라고도 불리며 N:N 통신 가능
  - 고가, 빠름

## 무선 LAN
- WiFi 랜카드도 MAC주소를 가지고 있음
- 유선이 아닌 무선으로 AP(Access Point)에 연결

## 유선+무선 복합 LAN
- 유선과 무선을 함께 사용하는 네트워크 구성


# 소켓(Socket)

실제 데이터 전송은 수신/송신 컴퓨터의 프로세스와 프로세스 사이에서 이루어진다.

## 전송 종단점(End Point)
- 컴퓨터 내부에서 동시에 실행되는 프로세스를 구분하기 위해 포트번호 사용
- 전체적으로 프로세스의 위치를 정확히 나타내기 위해 필요한 정보:
  - 사용 프로토콜
  - 송신자 주소와 포트번호
  - 수신자 주소와 포트번호

## 정확한 주소 구성
- 송신자 IP 주소
- 송신 포트
- 수신자 IP 주소  
- 수신 포트

개별 프로그램은 포트번호로 구분된다.
