9번 문제: 앞의 Person 클래스에서 새로운 Person 객체가 생성될 때 ID를 1001번지부터 순차적으로 자동으로 부여할 수 있도록 Employee 클래스를 포함해서 변경하시오. 새로운 Employee 객체를 만들어 getId()를 호출할 때 Id가 중복되면 안된다.

```python
class Person:
    def __init__(self, name, age): 
        self.name = name 
        self.age = age
        self.__id = ++count
    def getId(self): 
        print(self.__id) 
    def getAge(self): 
        print(self.age)
```


````python
class Person:
    # 클래스 변수: 다음에 부여할 ID 번호 (1001부터 시작)
    next_id = 1001
    
    def __init__(self, name: str, age: int): 
        self.name = name 
        self.age = age
        # 자동으로 ID 부여 (중복 방지)
        self.__id = Person.next_id
        Person.next_id += 1  # 다음 ID를 위해 증가
        print(f"Person '{name}'이 생성되었습니다. (ID: {self.__id})")
    
    def getId(self) -> int: 
        return self.__id
    
    def getAge(self) -> int: 
        return self.age
    
    def get_info(self) -> str:
        return f"이름: {self.name}, 나이: {self.age}, ID: {self.__id}"


# Person 클래스를 상속받는 Employee 클래스 정의
class Employee(Person):
    def __init__(self, name: str, age: int, department: str = "일반"):
        # 부모 클래스의 생성자 호출 (ID는 자동으로 부여됨)
        super().__init__(name, age)
        self.department = department
        print(f"직원 '{name}'이 등록되었습니다. (부서: {department})")
    
    def get_department(self) -> str:
        return self.department
    
    def get_employee_info(self) -> str:
        return f"직원 - {self.get_info()}, 부서: {self.department}"


print("--- 9번 문제 실행 ---")
print(f"초기 다음 ID: {Person.next_id}")

# Person 객체들 생성 (ID 자동 부여)
print("\n--- Person 객체 생성 ---")
person1 = Person("홍길동", 25)
person2 = Person("김철수", 30)
person3 = Person("이영희", 28)

print(f"\n현재 다음 ID: {Person.next_id}")

# Employee 객체들 생성 (ID 자동 부여)
print("\n--- Employee 객체 생성 ---")
emp1 = Employee("박민수", 35, "개발팀")
emp2 = Employee("정수진", 27, "마케팅팀")
emp3 = Employee("최동현", 32, "영업팀")

print(f"\n현재 다음 ID: {Person.next_id}")

# ID 중복 확인
print("\n--- ID 중복 확인 ---")
all_people = [person1, person2, person3, emp1, emp2, emp3]
ids = [person.getId() for person in all_people]
print(f"모든 ID: {ids}")
print(f"중복된 ID가 있는가? {len(ids) != len(set(ids))}")

# 정보 출력
print("\n--- 모든 객체 정보 ---")
for person in all_people:
    if isinstance(person, Employee):
        print(person.get_employee_info())
    else:
        print(person.get_info())
````
