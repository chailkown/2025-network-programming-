4번 문제: Articles 클래스를 작성하라. 이 클래스는 각각 제품의 이름, 재고수량, 단가를 나타내는 name, stock, price 인스턴스 변수를 가지고 있다. offer_price() 메소드는 구매 수량을 인수로 전달받아 총가격을 반환한다. 구매수량이 1~10개 이하이면 정상가격을 반환하고, 11~50개이면 5% 할인된 가격을 반환한다. 또 51개 이상이면 10% 할인해준다. purchase() 메소드는 구매수량을 인수로 전달받아 재고수량을 구매수량만큼 감소한다.


```python
class Articles:
    def __init__(self, name: str, stock: int, price: int):
        # 제품 정보 초기화
        self.name = name      # 제품명
        self.stock = stock    # 재고수량
        self.price = price    # 단가
        print(f"제품 '{name}'이 등록되었습니다. (재고: {stock}개, 단가: {price}원)")
    
    def offer_price(self, quantity: int) -> float:
        # 구매 수량에 따른 할인가격 계산
        if quantity <= 0:
            print("구매 수량은 1개 이상이어야 합니다.")
            return 0
        
        total_price = self.price * quantity
        
        if 1 <= quantity <= 10:
            # 1~10개: 정상가격
            discount_rate = 0
            print(f"정상가격 적용: {quantity}개 × {self.price}원 = {total_price}원")
        elif 11 <= quantity <= 50:
            # 11~50개: 5% 할인
            discount_rate = 0.05
            total_price = total_price * (1 - discount_rate)
            print(f"5% 할인 적용: {quantity}개 × {self.price}원 × 0.95 = {total_price:.0f}원")
        else:
            # 51개 이상: 10% 할인
            discount_rate = 0.10
            total_price = total_price * (1 - discount_rate)
            print(f"10% 할인 적용: {quantity}개 × {self.price}원 × 0.90 = {total_price:.0f}원")
        
        return total_price
    
    def purchase(self, quantity: int) -> bool:
        # 구매 처리 (재고 감소)
        if quantity <= 0:
            print("구매 수량은 1개 이상이어야 합니다.")
            return False
        
        if self.stock < quantity:
            print(f"재고 부족: 현재 재고 {self.stock}개, 요청 수량 {quantity}개")
            return False
        
        self.stock -= quantity
        print(f"구매 완료: {quantity}개 구매, 남은 재고: {self.stock}개")
        return True
    
    def get_info(self) -> str:
        # 제품 정보 반환
        return f"제품명: {self.name}, 재고: {self.stock}개, 단가: {self.price}원"


# 테스트 코드
print("--- Articles 클래스 테스트 ---")
mask = Articles("mask", 10000, 1000)

# 가격 계산 테스트
print("\n--- 가격 계산 테스트 ---")
mask.offer_price(5)    # 정상가격
mask.offer_price(25)   # 5% 할인
mask.offer_price(60)   # 10% 할인

# 구매 테스트
print("\n--- 구매 테스트 ---")
mask.purchase(100)     # 정상 구매
mask.purchase(20000)   # 재고 부족

# 현재 정보 확인
print(f"\n현재 정보: {mask.get_info()}")
```

```python
# 실행 결과
mask = Articles("mask", 10000, 1000)
mask.offer_price(5)
mask.offer_price(25)
mask.offer_price(60)
mask.purchase(100)
mask.purchase(20000)
print(mask.get_info())
```
