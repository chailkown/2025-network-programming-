8번 문제: 동물(Animal)을 울음을 추상메소드로 하는 추상클래스로 정의하고,
• 개(Dog)와 고양이(Cat)를 구체적인 클래스로 정의한 후,
• 각 동물의 울음소리를 출력하는 프로그램을 작성하시오.
• 각 동물 객체는 자신의 이름을 속성으로 가지며,
• 생성자에서 이름을 초기화합니다.
• 추가적으로 Animal 클래스는 현재 생성된 동물 객체의 수를 클래스변수로 가지고,
• 정적메소드로 현재 동물 객체의 수를 반환하는 메소드를 구현
• 객체가 생성(__init__)될 때 객체의 수를 증가하고 삭제(__del__(self)) 될 때 객체의 수를 감소

- 테스트
• 개나 고양이 객체들을 리스트로 생성하고 개나 Animal 변수로 받아 동적바인딩이 되는지 확인하시오.
• 모든 객체들이 울게 하는 함수를 작성하여 테스트하시오.


```python
from abc import ABC, abstractmethod

# 추상 클래스 Animal 정의
class Animal(ABC):
    # 클래스 변수: 현재 생성된 동물 객체의 수
    animal_count = 0
    
    def __init__(self, name: str):
        self.name = name
        # 객체 생성 시 카운트 증가
        Animal.animal_count += 1
        print(f"동물 '{name}'이 생성되었습니다. (총 동물 수: {Animal.animal_count})")
    
    @abstractmethod
    def make_sound(self):
        # 추상 메소드: 울음소리 (자식 클래스에서 반드시 구현해야 함)
        pass
    
    @staticmethod
    def get_animal_count() -> int:
        # 정적 메소드: 현재 동물 객체 수 반환
        return Animal.animal_count
    
    def __del__(self):
        # 객체 삭제 시 카운트 감소
        Animal.animal_count -= 1
        print(f"동물 '{self.name}'이 삭제되었습니다. (남은 동물 수: {Animal.animal_count})")


# 구체 클래스 Dog 정의
class Dog(Animal):
    def __init__(self, name: str):
        super().__init__(name)
    
    def make_sound(self):
        print(f"{self.name}: 멍멍!")


# 구체 클래스 Cat 정의
class Cat(Animal):
    def __init__(self, name: str):
        super().__init__(name)
    
    def make_sound(self):
        print(f"{self.name}: 야옹!")


# 모든 동물이 울게 하는 함수 (다형성 테스트)
def make_all_animals_sound(animals: list):
    print("\n--- 모든 동물이 울기 시작 ---")
    for animal in animals:
        animal.make_sound()


print("--- 8번 문제 실행 ---")
print(f"초기 동물 수: {Animal.get_animal_count()}")

# 동물 객체들 생성
print("\n--- 동물 객체 생성 ---")
dog1 = Dog("바둑이")
cat1 = Cat("나비")
dog2 = Dog("멍멍이")
cat2 = Cat("야옹이")

print(f"\n현재 동물 수: {Animal.get_animal_count()}")

# 동적 바인딩 테스트
print("\n--- 동적 바인딩 테스트 ---")
animals = [dog1, cat1, dog2, cat2]  # Animal 타입으로 리스트에 저장
make_all_animals_sound(animals)

# 개별 동물 울음 테스트
print("\n--- 개별 동물 울음 테스트 ---")
dog1.make_sound()
cat1.make_sound()

# 객체 삭제 테스트
print("\n--- 객체 삭제 테스트 ---")
del dog1
del cat1
print(f"삭제 후 동물 수: {Animal.get_animal_count()}")
```

```python
# 실행 결과
print(f"초기 동물 수: {Animal.get_animal_count()}")

dog1 = Dog("바둑이")
cat1 = Cat("나비")
dog2 = Dog("멍멍이")
cat2 = Cat("야옹이")

print(f"현재 동물 수: {Animal.get_animal_count()}")

animals = [dog1, cat1, dog2, cat2]
make_all_animals_sound(animals)

dog1.make_sound()
cat1.make_sound()

del dog1
del cat1
print(f"삭제 후 동물 수: {Animal.get_animal_count()}")
```
